```python
import pandas as pd
df = pd.read_csv('/content/data.csv', encoding='latin1')
df.head()
```

```python
x = df["no2"].dropna().values
print("Total NO2 samples:", len(x))
```

```python
df["no2"] = pd.to_numeric(df["no2"], errors="coerce")
x = df["no2"].dropna().values
print("Clean NO2 samples:", len(x))
```

```python
import numpy as np
r = 102303147
a_r = 0.05 * (r % 7)
b_r = 0.3 * (r % 5 + 1)
print("a_r =", a_r)
print("b_r =", b_r)
```

```python
z = x + a_r * np.sin(b_r * x)
print("z sample:", z[:10])
print("z min, max:", z.min(), z.max())
```

```python
mu = np.mean(z)
print("mu =", mu)
```

```python
sigma2 = np.var(z)
lam = 1 / (2 * sigma2)
print("lambda =", lam)
```

```python
import math
c = math.sqrt(lam / math.pi)
print("c =", c)
```

```python
print("===== SUBMISSION VALUES =====")
print("lambda =", lam)
print("mu     =", mu)
print("c      =", c)
```
